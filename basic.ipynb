```python

def oddfact(n): 
    fact=1
    for i in range (1,n+1,2):
        fact=fact*i
    return fact

def evenfact(n):
    fact=1
    for i in range (2,n+1,2):
      fact=fact*i
    return fact


n = int(input("Enter the number to find the factorial : "))

if n%2==0:
   print(oddfact(n)) 
    
else:
    print(evenfact(n))
```

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f    

n= int(input("enter the value of n : "))
r= int(input("enter the value of r : "))

a=fact(n)
b=fact(n-r)
c=a/(b*r)
print(c)
```
